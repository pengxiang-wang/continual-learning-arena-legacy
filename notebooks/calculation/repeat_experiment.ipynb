```python
import pandas as pd
```

```python

metric = "test_metrics"
# metric = "forgetting_ratio"

# csv_path_list = [f"../../logs_saved/s20/HAT_1/runs/s{i}/csv/{metric}.csv" for i in range(1,11)]


csv_path_list = [f"../../logs_saved/l50/ada_cons_alpha_all/runs/l{i}/csv/{metric}.csv" for i in range(1,11)]
```

```python
dfs = [pd.read_csv(csv_path) for csv_path in csv_path_list]

mean_result = pd.concat(dfs, axis=0).groupby(level=0).mean()
variance_result = pd.concat(dfs, axis=0).groupby(level=0).std()


print(mean_result)
mean_result.to_csv(f"{metric}_mean.csv")


print(variance_result)
variance_result.to_csv(f"{metric}_std.csv")
```
