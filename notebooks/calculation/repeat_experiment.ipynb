```python
import pandas as pd
import os
```

```python

# metric = "test_metrics"
# metric = "test_forgetting_ratio"
# metric = "test_fm"
# metric = "test_bwt"
# metric = "test_fwt"
metric = "test_im"

# csv_path_list = [f"../../logs_saved/s20/HAT_1/runs/s{i}/csv/{metric}.csv" for i in range(1,11)]

expr_names = [
    "Finetuning_1",
    "LwF_1",
    "EWC_1", 
    "HAT_1", 
    "ada", 
    "ada_ave", 
    "ada_cons_1",
        "ada_cons_alpha",

    "ada_prob", 
    "ada_ave_prob", 
    "ada_cons_alpha_all",
    "ada_random",
    "ada_random_all",
        "ada_random_all",

    "ada_reg_1",
    "ada_reg_0",
        "ada_reg_09",

    "ada_sum_1",
    "ada_sum_t",
    
    "Random_1",
    "Reference_1",
    # "Joint_1",
]


```

```python
for expr in expr_names:

    output_path = f"../../logs_saved/pmnist/{expr}/runs"
    
    if os.path.exists(output_path):

        csv_path_list = [f"{output_path}/l{i}/csv/{metric}.csv" for i in range(1,11)]


        dfs = [pd.read_csv(csv_path) for csv_path in csv_path_list]

        mean_result = pd.concat(dfs, axis=0).groupby(level=0).mean()
        variance_result = pd.concat(dfs, axis=0).groupby(level=0).std()


        print(mean_result)
        mean_result.to_csv(f"{output_path}/{metric}_mean.csv", index=False)
        

        print(variance_result)
        variance_result.to_csv(f"{output_path}/{metric}_std.csv", index=False)
```
