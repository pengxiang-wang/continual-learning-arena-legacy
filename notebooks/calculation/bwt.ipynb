```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
 
expr_names = [
    "Finetuning_1",
    "LwF_1",
    "EWC_1", 
    "HAT_1", 
    "ada", 
    "ada_ave", 
    "ada_cons_1",
    "ada_cons_alpha",
    "ada_prob", 
    "ada_ave_prob", 
    "ada_cons_alpha_all",
    "ada_random",
    "ada_random_all",
    "ada_reg_1",
    "ada_reg_0",
    "ada_reg_09",
    "ada_sum_1",
    "ada_sum_t",
    "Random_1",
    # "Joint_1",
    ]
 
 
expr_names = [
    "EWC_4",
    ]
 
 
  
 
 # read csv

for n in expr_names:
    for i in range(1,11):
        print(n)
        
        csv_path = f"../../logs/{n}/runs/{i}/csv/test_metrics.csv"

        output_csv_path = f"../../logs/{n}/runs/{i}/csv/test_bwt.csv"

        df = pd.read_csv(csv_path)

        task = df["task"]
        task_column = pd.Series(task, name="task") 

        # parse df
        for column in df.columns:
            if "test/acc/task" not in column:
                df.drop(column, axis=1, inplace=True)
                
        bwt = []

        for k in range(df.shape[0]):
            if k == 0:
                bwt.append(np.nan)
            else:
                # print(f"test/acc/task{i}")            
                bwt_k = 0
                for j in range(k):
                    bwt_k += (df.at[k, f"test/acc/task{j}"] - df.at[j, f"test/acc/task{j}"])
                    # print(forgetting_ratio_df.iloc[i, j])
                bwt_k /= k
                bwt.append(bwt_k)

        bwt_column = pd.Series(bwt, name="bwt") 


        bwt_df = pd.concat([task, bwt_column], axis=1)       

        bwt_df.to_csv(output_csv_path, index=False)
```
