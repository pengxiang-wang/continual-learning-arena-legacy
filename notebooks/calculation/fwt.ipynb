```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
 
expr_names = [
    "Finetuning_1",
    "LwF_1",
    "EWC_1", 
    "HAT_1", 
    "ada", 
    "ada_ave", 
    "ada_cons_1",
    "ada_cons_alpha",
    "ada_prob", 
    "ada_ave_prob", 
    "ada_cons_alpha_all",
    "ada_random",
    "ada_random_all",
    "ada_reg_1",
    "ada_reg_0",
    "ada_reg_09",
    "ada_sum_1",
    "ada_sum_t",
    "Random_1",
    # "Joint_1",
    ]



expr_names = [
    "EWC_4",
    ]
 
 
  
 
 
# read csv

for n in expr_names:
    for i in range(1,11):
        print(n)
        
        csv_path = f"../../logs/{n}/runs/{i}/csv/test_metrics.csv"

        ref_csv_path = "../../logs/Reference_4/runs/1/csv/test_metrics.csv"
        
        output_csv_path = f"../../logs/{n}/runs/{i}/csv/test_fwt.csv"

        df = pd.read_csv(csv_path)
        ref_df = pd.read_csv(ref_csv_path)


        task = df["task"]
        task_column = pd.Series(task, name="task") 

        # parse df
        for column in df.columns:
            if "test/acc/task" not in column:
                df.drop(column, axis=1, inplace=True)
                
        fwt = []

        for k in range(df.shape[0]):
            if k == 0:
                fwt.append(np.nan)
            else:
                # print(f"test/acc/task{i}")            
                fwt_k = 0
                for j in range(1,k+1):
                    fwt_k += (df.at[j, f"test/acc/task{j}"] - ref_df.at[k, f"test/acc/task{j}"])
                    # print(forgetting_ratio_df.iloc[i, j])
                fwt_k /= k
                fwt.append(fwt_k)
                
        print(fwt)

        fwt_column = pd.Series(fwt, name="fwt") 


        fwt_df = pd.concat([task, fwt_column], axis=1)       

        fwt_df.to_csv(output_csv_path, index=False)
```
